```python
import pandas as pd

df = pd.read_csv('/content/sample_data/presidents.csv')
df.head(10)
```

```python
def splitname(row):
  row['First']= row['President '].split(" ")[0]
  row['Last']= row['President '].split(" ")[-1]
  return row

df = df.apply(splitname, axis = 'columns')
df.head()
```

```python
del(df['First'])
del(df['Last'])
```

```python
pattern = "(^[\w]*)(?:.* )([\w]*$)"
df['President '].str.extract(pattern).head()
```

```python
pattern = "(?P<First>^[\w]*)(?:.* )(?P<Last>[\w]*$)"

names = df['President '].str.extract(pattern).head()
names
```

```python
df['First']=names['First']
df['Last']=names['Last']
df.head()
```

```python
df['Took office '] = pd.to_datetime(df['Took office '])
df.head()
```
