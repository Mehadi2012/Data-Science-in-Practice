```python
# Import packages
from urllib.request import urlopen,Request

# Specify the url
url = "https://google.com"

# This packages the request: request

request = Request(url)
# Sends the request and catches the response: response
response = urlopen(request)

# Print the datatype of response
print(type(response))

# close the response!
response.close()
```

```python
# Import packages
from urllib.request import urlopen, Request

# Specify the url
url = "https://google.com"

# This packages the request
request = Request(url)

# Sends the request and catches the response: response

response = urlopen(request)
# Extract the response: html

html = response.read()
# Print the html
print(html)

# close the response!
response.close()
```

```python
# Import package

import requests

# Specify the url: url

url = "http://google.com"

# Packages the request, send the request and catch the response: r

r = requests.get(url)

# Extract the response: text

text = r.text

# Print the html

print(text)
```
