```python
from urllib.request import urlopen,Request

url = "http://bracu.ac.bd"

request = Request(url)

response = urlopen(request)

html = response.read()

print(html)

response.close()
```

```python
from urllib.request import urlopen,Request

url = 'http://google.com'

r = Request(url)

resp = urlopen(r)

t_html = type(resp)
print(t_html)

resp.close()
```
